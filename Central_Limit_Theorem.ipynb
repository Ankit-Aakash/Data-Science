Central Limit Theorem states that, given a population having any distribution, if we take some n samples (n >= 30) then, distribution of the mean of the sample will tend to normal distribution. 

Moreover, mean and standard deviatioin of the samples mean will be equal to mean of population and std(population)/sqrt(number of samples) respectively.

```python
import numpy as np
import numpy.random as ran
import seaborn as sns

population_size = 10000
population = ran.rand(population_size)
sns.displot(population) #We can see that the distribution of population is not normal distribution
```

```python
samplesize = 100
numofsamples = 10000
```

```python
#Take 100 samples of sample size 10 and calculate their mean
sample_mean=[]
for i in range(0,numofsamples):
  sample = ran.choice(population,size = samplesize) #Random choices from population of size samplesize
  sample_mean.append(np.mean(sample)) #Store the mean of the chosen sample

sns.displot(sample_mean)
```

```python
print('Mean of Population :'+ str(np.mean(population)))
print('Standard Deviation of the population: '+ str(np.std(population)))
```

```python
print('Mean of Sample_Mean :'+ str(np.mean(sample_mean)))
print('Standard Deviation of the Sample_Mean: '+ str((np.std(population)/np.sqrt(numofsamples))))
```
